## Foundry 포털에서의 모니터링

- Foundry 포털에 접속하여 배포된 에이전트의 실시간 메트릭 및 상태를 확인하세요.
- 내장된 대시보드를 사용하여 신속하게 인사이트를 얻을 수 있습니다.

![](../assets/05-tracing-resourceusage.png)

## Application Insights Dashboards

- 애플리케이션 인사이트를 설정하여 애플리케이션의 원격 측정 데이터를 수집합니다.
- 사용자 지정 대시보드를 생성하여 성능, 오류 및 사용량을 시각화합니다.

![](../assets/05-tracing-applicationinsights.png)

## Tracing with OpenTelemetry

- 분산 추적을 위해 애플리케이션에 OpenTelemetry를 구현합니다.
- 추적 결과를 Azure Monitor 또는 기타 관찰 플랫폼으로 내보냅니다.

![](../assets/05-tracing-single-trace.png)
