## AI 에이전트란 ​​무엇인가?

**AI 에이전트**는 대규모 언어 모델(LLM)을 사용하여 목표를 이해하고, 결정을 내리고, 도구를 호출하고, 최소한의 사람 개입으로 작업을 완료하는 자율 시스템입니다.

**주요 차이점:**
- **AI 어시스턴트**: 사람과의 상호 작용을 지원하고, 질문에 답하고, 정보를 제공합니다.
- **AI 에이전트**: 자율적으로 행동하고, 도구를 호출하고, 목표 달성을 위한 워크플로에 참여합니다.

**예시**: 고객이 Cora에게 "어떤 페인트가 있나요?"라고 물으면 에이전트는 다음과 같은 작업을 수행합니다.
1. 질문을 이해합니다(언어 모델 사용).
2. 어떤 도구를 사용할지 결정합니다(제품 검색, 재고 확인).
3. 정보를 검색합니다(자율적으로 도구 호출).
4. 유용한 답변을 생성합니다(결과 종합).

에이전트는 단순히 말로만 하는 것이 아니라 실제로 *일을 할 수 있는* 어시스턴트라고 생각하면 됩니다.

**에이전트의 주요 특징:**
- **자율성**: 정해진 제약 조건 내에서 독립적으로 의사 결정을 내립니다.
- **도구 활용**: 함수 호출, API 사용, 데이터베이스 검색 등을 할 수 있습니다.
- **추론**: LLM 기능을 활용하여 계획을 수립하고 문제를 해결합니다.
- **목표 지향성**: 특정 목표 달성을 위해 노력합니다.

## AI 에이전트의 세 가지 핵심 구성 요소

모든 AI 에이전트는 다음 세 가지 필수 구성 요소로 이루어져 있습니다.

### 1. 모델 (LLM)

**모델**은 언어 이해 및 추론 기능을 제공합니다.

**주요 기능:**
- 자연어 입력을 이해합니다.
- 사람과 유사한 응답을 생성합니다.
- 사용할 도구를 추론합니다.
- 여러 소스에서 정보를 종합합니다.

**모델 예시:**
- `gpt-4o` - 고급 추론, 멀티모달 기능
- `gpt-4o-mini` - 빠르고 효율적이며 비용 효율적입니다.
- `gpt-4.1` - 강력한 범용 성능

**모델 선택:**
- **복잡한 추론**: `gpt-4o` 또는 `gpt-4.1` 사용
- **대용량 처리 및 빠른 속도가 중요**: `gpt-4o-mini` 사용
- **멀티모달 요구 사항**(이미지, 오디오): `gpt-4o` 사용

### 2. 명령어

**지침**은 에이전트의 성격, 행동, 제약 조건 및 목표를 정의합니다.

**포함되는 내용:**
- 역할 및 성격 (예: "당신은 친절한 하드웨어 스토어 직원입니다.")
- 응답 스타일 (예: "정중하고 전문적인 태도를 유지하세요.")
- 제약 조건 (예: "사실에 근거한 제품 정보만 제공하세요.")
- 지침 (예: "모르는 정보는 모른다고 솔직하게 말하세요. 지어내지 마세요.")

**코라 예시:**
```python
instructions = """
당신은 Zava 하드웨어 스토어의 친절한 고객 서비스 직원 Cora입니다.

당신의 역할:
- 고객의 제품 관련 질문에 정중하고 정확하게 답변합니다.
- 고객이 집 꾸미기에 필요한 제품을 찾도록 돕습니다.
- 요청 시 SKU, 가격, 재고 정보를 제공합니다.
- 제품 정보를 지어내지 마세요. 제공된 도구만 ​​사용하세요.

당신의 어조:
- 친절하고 전문적인 어조
- 인내심 있고 도움이 되는 어조
- 명확하고 간결한 어조
"""
```

좋은 지침은 일반적인 챗봇과 전문적인 에이전트를 구분 짓는 중요한 요소입니다.

### 3. 도구

**도구**는 에이전트의 기능을 언어 이해를 넘어 확장하여 에이전트가 행동을 취하고 정보를 검색할 수 있도록 합니다.

**일반적인 도구 유형:**
- **함수 호출**: 사용자 지정 Python 함수 실행
- **파일 검색**: 업로드된 문서 검색
- **코드 인터프리터**: Python 코드 동적 실행
- **API 호출**: 외부 시스템 및 데이터베이스 액세스
- **지식 기반**: 벡터 데이터베이스, Azure AI 검색 쿼리

**Cora용 도구 예시:**
- `get_product_info(sku)` - 제품 세부 정보 조회
- `check_inventory(sku)` - 현재 재고량 조회
- `search_products(query)` - 검색 조건과 일치하는 제품 찾기
- `calculate_discount(customer_tier, cart_value)` - 개인 맞춤 가격 계산

**도구 작동 방식:**
1. 에이전트가 고객 질문을 수신합니다.
2. 어떤 도구를 사용할지 모델링합니다.
3. 에이전트가 적절한 매개변수를 사용하여 도구를 호출합니다.
4. 도구가 결과를 반환합니다.
5. 에이전트가 결과를 사용하여 응답을 구성합니다.

도구는 에이전트를 단순한 대화 전달자에서 능동적인 행동 전달자로 변화시킵니다.

## Key Concepts: Threads, Messages, and Runs

Azure AI Foundry에서 에이전트를 사용하려면 이러한 개념을 이해하는 것이 필수적입니다.

### Threads

**스레드**란 사용자와 에이전트 간의 단일 대화 세션을 나타냅니다.

**채팅 창 또는 대화 기록**이라고 생각하면 됩니다.

**구성 요소:**
- 사용자와 상담원 간에 주고받은 모든 메시지
- 여러 턴에 걸쳐 유지되는 컨텍스트
- 대화에 대한 메타데이터

**예시:**
```python
# Create a new thread (new conversation)
thread = client.agents.create_thread()

# Later, use the same thread to continue the conversation
thread_id = thread.id
```

**사용 사례:**
- 고객 지원 세션당 하나의 스레드
- 여러 턴으로 진행되는 애플리케이션에서 사용자당 하나의 스레드
- 주제/대화별로 별도의 스레드

### Message

**메시지**는 스레드 내에서 사용자 또는 상담원이 주고받는 단일 교환입니다.

**Message types:**
- **사용자 메시지**: 고객의 질문 또는 요청
- **어시스턴트 메시지**: 상담원이 생성한 응답

**Example:**
```python
# User sends a message
message = client.agents.create_message(
    thread_id=thread.id,
    role="user",
    content="What paint do you have for exterior wood?"
)

# Agent responds (created during Run)
# Assistant message is automatically added to thread
```

**Important**: 메시지는 대화 맥락을 유지하면서 시간 순서대로 저장됩니다.

### Run

**실행**이란 에이전트가 다음 작업을 수행하는 처리 주기입니다.

1. 최신 사용자 메시지를 읽습니다.
2. 응답 방법에 대한 이유를 파악합니다.
3. 필요한 경우 도구를 호출합니다.
4. 응답을 생성합니다.
5. 응답을 스레드에 추가합니다.

**간단히 말해,** 에이전트가 "생각"하고 조치를 취하는 과정입니다.

**Example:**
```python
# Start a run (agent processes the thread)
run = client.agents.create_run(
    thread_id=thread.id,
    assistant_id=assistant.id
)

# Wait for run to complete
while run.status in ["queued", "in_progress"]:
    run = client.agents.get_run(thread_id=thread.id, run_id=run.id)
    time.sleep(1)

# Run completed - agent's response is now in the thread
```

**Run statuses:**
- `queued` - Waiting to start
- `in_progress` - Agent is processing
- `requires_action` - Needs tool approval (if required)
- `completed` - Finished successfully
- `failed` - Encountered an error

### How They Work Together

```
1. 스레드 생성
   ↓
2. 사용자 사용자가 스레드에 메시지를 추가
   ↓
3. 실행 생성(에이전트 처리)
   ↓
4. 실행 완료 → 어시스턴트 메시지가 스레드에 추가됨
   ↓
5. 사용자가 다른 메시지를 추가
   ↓
6. 다른 실행 생성
   ↓
   (주기가 계속됨...)
```

각 실행은 스레드의 이전 메시지를 모두 볼 수 있으므로 대화 전체의 맥락을 유지합니다.

## Agent Architecture Patterns

### Single-Agent Pattern

**Use when:** One agent can handle all tasks effectively.

```
Customer Question
       ↓
   [Cora Agent]
   - Model: GPT-4o-mini
   - Tools: Product search, inventory check
   - Instructions: Be helpful, accurate
       ↓
   Response
```

**Pros:** Simple, easy to manage, fast
**Cons:** Can become complex if handling too many responsibilities

### Multi-Agent Pattern

**Use when:** Different specialized agents handle different aspects.

```
Customer Question
       ↓
[Routing Agent] → determines intent
       ↓
    ┌─────┴─────┬─────────┬────────┐
    ↓           ↓         ↓        ↓
[Product    [Inventory  [Discount  [Support
 Agent]      Agent]      Agent]     Agent]
    ↓           ↓         ↓        ↓
         Combined Response
```

**Pros:** Specialized expertise, easier to maintain, can optimize each agent
**Cons:** More complex orchestration, higher latency

### Agentic Workflow Pattern

**Use when:** Tasks require sequential steps or decision trees.

```
Customer: "I need paint for my deck"
    ↓
Step 1: [Understanding Agent] → Deck = exterior wood surface
    ↓
Step 2: [Product Agent] → Find exterior wood paint products
    ↓
Step 3: [Recommendation Agent] → Suggest based on weather, budget
    ↓
Step 4: [Discount Agent] → Apply customer tier pricing
    ↓
Final Response
```

**Pros:** Complex reasoning, step-by-step validation
**Cons:** Higher cost, slower response time

## 에이전트 사용 시점 vs. 단순 프롬프트 사용 시점

### 단순 프롬프트 사용 시점:

- ✅ 작업이 간단한 경우 (분류, 요약, 질문 및 답변)
- ✅ 외부 데이터가 필요 없는 경우 (프롬프트에 모든 컨텍스트 포함)
- ✅ 별도의 작업이 필요하지 않은 경우 (텍스트 생성만)
- ✅ 지연 시간이 매우 중요한 경우 (단일 API 호출)

**예시:** "이 제품 설명을 두 문장으로 요약해 주세요."

### 에이전트 사용 시점:

- ✅ 외부 정보 검색이 필요한 경우 (데이터베이스, API, 검색)
- ✅ 여러 단계의 추론이 필요한 경우 (계획 → 실행 → 검증)
- ✅ 특정 작업을 수행해야 하는 경우 (레코드 업데이트, 이메일 전송, 구매)
- ✅ 복잡한 워크플로우 (만약 ~라면 ~이다)
- ✅ 대화 컨텍스트를 유지해야 하는 경우

**예시:** "내 야외 데크에 가장 적합한 페인트를 찾아주고, 재고 여부를 확인하고, 할인율을 계산해 주세요."

### Hybrid Approach:

Start simple, add complexity as needed:

```
Level 1: Simple prompt (no tools)
    ↓
Level 2: Prompt + RAG (retrieve knowledge)
    ↓
Level 3: Agent with tools (retrieve + act)
    ↓
Level 4: Multi-agent system (orchestration)
```

**일반적인 규칙:** 문제를 효과적으로 해결하는 가장 간단한 접근 방식을 사용하십시오.

## Best Practices for Building Agents

### 1. Write Clear Instructions

```python
# Good - specific and actionable
instructions = """
You are Cora, a Zava Hardware Store assistant.
When customers ask about products:
1. Use get_product_info to retrieve accurate details
2. Always include SKU, price, and description
3. If out of stock, suggest alternatives
4. Never fabricate information
"""

# Less effective - vague
instructions = "You are a helpful assistant. Be nice."
```

### 2. 도구를 현명하게 선택하세요

**도구가 너무 적으면:** 상담원이 작업을 완료할 수 없습니다.
**도구가 너무 많으면:** 상담원이 어떤 도구를 사용해야 할지 혼란스러워합니다.

**지침:** 필수 도구 3~5개로 시작하고 필요에 따라 추가하세요.

### 3. Test Tool Descriptions

도구에는 상담원이 언제 사용해야 할지 판단하는 데 도움이 되는 설명이 있어야 합니다.

```python
# Good - clear when to use
{
    "name": "get_product_info",
    "description": "SKU로 제품 세부 정보를 조회합니다. 고객이 특정 제품, 가격 또는 기능에 대해 문의할 때 사용하세요."
}

# Unclear
{
    "name": "get_product_info", 
    "description": "제품 데이터를 가져옵니다"
}
```

### 4. Handle Failures Gracefully

```python
# In instructions:
"""
도구 호출이 실패할 경우:
1. 고객에게 사과하세요
2. 무엇이 잘못되었는지 간단하게 설명하세요
3. 가능하다면 대안을 제시하세요
4. 기술적인 오류 세부 정보를 절대 공개하지 마세요
"""
```

### 5. 모니터링 및 반복

- 가장 자주 사용되는 도구를 추적하세요
- 일반적인 오류 패턴을 파악하세요
- 대화 로그를 검토하여 개선 기회를 찾으세요
- 평가 지표를 사용하여 품질을 측정하세요

## Terminology Quick Reference

| Term | Simple Definition |
|------|-------------------|
| **Agent** | LLM(언어 학습 모델)을 사용하여 이해하고, 결정하고, 행동하는 자율 AI 시스템 |
| **Assistant** | Azure AI Foundry에서 에이전트를 부르는 다른 이름 |
| **Model** | 언어 이해를 제공하는 LLM(예: GPT-4o) |
| **Instructions** | 에이전트의 성격, 동작 및 제약 조건을 정의하는 텍스트 |
| **Tools** | 에이전트가 데이터를 검색하거나 작업을 수행하기 위해 호출할 수 있는 함수 |
| **Thread** | 사용자와 에이전트 간의 대화 세션 |
| **Message** | 스레드(사용자 또는 어시스턴트) 내의 단일 교환 |
| **Run** | 에이전트가 생각하고 응답하는 처리 주기 |
| **Orchestration** | 대화 및 도구 호출 흐름 관리 |
| **Azure AI Agent Service** | 에이전트 구축 및 운영을 위한 관리형 플랫폼 |
| **Function Calling** | 상담원이 매개변수를 사용하여 도구를 호출할 수 있는 기능 |
| **Context Window** | 모델이 한 번에 처리할 수 있는 텍스트 양 |
| **Token** | 청구에 사용되는 텍스트 단위(약 4자) |
